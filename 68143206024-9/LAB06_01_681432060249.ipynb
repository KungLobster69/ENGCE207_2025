```python
# Workshop 01:  K-Means
#   n_clusters  K
#  fit  predict  X  y_kmeans
#  make_blobs  300 , 4  (centers=4)  0.7 (cluster_std=0.7)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=42)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering on Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

#  K-Means Clustering  Elbow Method
# 1.  K-Means Clustering
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
axes[0].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

axes[1].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
axes[1].set_title('True Clusters')

# 2.  Elbow Method

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init ='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Synthetic Data')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('(WCSS)')
plt.grid(True)
plt.show()


```
