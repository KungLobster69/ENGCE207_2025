```python
import seaborn as sns 
import matplotlib.pyplot as plt

# โหลด dataset titanic
titanic = sns.load_dataset("titanic")

# --- กราฟที่ 3: Catplot ---
# สร้างตารางกราฟด้วย catplot
# - x='sex', y='survived': ดูความสัมพันธ์ระหว่างเพศกับอัตราการรอดชีวิต
# - col='pclass': แยกกราฟตามชั้นที่นั่ง (pclass)
# - kind='bar': ใช้แผนภูมิแท่งในการแสดงข้อมูล
sns.catplot(data=titanic, x='sex', y='survived', col='pclass', kind='bar')

# แสดงกราฟ
plt.show()

# --- กราฟที่ 4: Violin Plot ---
# กรองข้อมูลเฉพาะแถวที่มีอายุ
titanic_filtered = titanic.dropna(subset=['age'])

# สร้างกราฟ Violin Plot เพื่อดูการกระจายของอายุในแต่ละชั้นที่นั่ง และสถานะการรอดชีวิต
sns.violinplot(data=titanic_filtered, x='pclass', y='age', hue='survived', split=True)

# ตั้งชื่อหัวข้อของกราฟ
plt.title('Age Distribution by Class and Survival')

# แสดงกราฟ
plt.show()
```
