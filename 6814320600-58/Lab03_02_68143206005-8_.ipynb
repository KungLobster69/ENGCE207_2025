```python
import pandas as pd
import numpy as np

df = pd.read_csv('vgsales.csv')

#2.1 การเลือกคอลัมน์
print("\n2.1 เลือกแสดงข้อมูลเฉพาะคอลัมน์ Name, Platform, Year, และGlobal_Sales:")
selected_columns = df[['Name', 'Platform', 'Year', 'Global_Sales']]
print(selected_columns.head())

#2.2 การคัดกรองตามเงื่อนไข
print("\n2.2.1 คัดกรองเกมที่ลงในแพลตฟอร์ม 'Wii':")
wii_games = df[df['Platform'] == 'Wii']
print(wii_games.head())

print("\n2.2.2 คัดกรองเกม Action ที่มียอดขายทั่วโลก > 10ล้าน:")
action_games_high_sales = df[(df['Genre'] == 'Action') & (df['Global_Sales'] > 10)]
print(action_games_high_sales)

```

```python
print("\n2.2.3 คัดกรองเกมจาก 'Nintando' หรือ 'Sony Computer Entertainmet':")
major_plublishers = ['Nintendo' , 'Sony Computer Entertainment']
nintendo_or_sony_games = df[df['Publisher'].isin(major_plublishers)]
print(nintendo_or_sony_games.head)
```
