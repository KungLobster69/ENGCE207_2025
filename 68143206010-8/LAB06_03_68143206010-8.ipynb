```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# สร้างข้อมูลตัวอย่าง (ข้อมูลเทียม)
X_synth, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# ทำ K-Means Clustering
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans_synth = kmeans.fit_predict(X_synth)

# วาดกราฟเปรียบเทียบผล K-Means และกลุ่มจริง
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# กราฟผลลัพธ์จาก K-Means
axes[0].scatter(X_synth[:, 0], X_synth[:, 1], c=y_kmeans_synth, s=50, cmap='viridis')
axes[0].set_title('K-Means Clustering Result')

# กราฟกลุ่มจริง
axes[1].scatter(X_synth[:, 0], X_synth[:, 1], c=y_true, s=50, cmap='viridis')
axes[1].set_title('True Clusters')

plt.show()
```
