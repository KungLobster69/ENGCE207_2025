```python
import pandas as pd
import numpy as np

df = pd.read_csv('vgsales.csv')

print("--- ส่วนที่ 3: การจัดขอข้อมูลที่ขาดหาย ---")

print("\n3.1 จำนวนข้อมูลที่ขาดหายในแต่ละคอลัม (ก่อนจัดการ):")
print(df.isnull().sum())

print("\n3.2 ลบแถวที่คอลัมน์ 'Year' มีค่าว่าง:")
df.dropna(subset=['Year'], inplace=True)

print("\n3.3 แทนที่คำว่างใน 'Publisher' ด้วย 'Unknown':")
df['Publisher'].fillna('Unknown', inplace=True)

print("\n3.4 จำนวนข้อมูลที่ขาดหายในแต่ละคอลัมน์ (หลังจัดการ):")
print(df.isnull().sum())
print("จัดการข้อมูลที่ขาดหายเรียบร้อยแล้ว")

print("\n--- สิ้นสุดส่วนที่ 3 ---\n" + "="*50 + "\n")
```
