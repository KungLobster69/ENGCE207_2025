```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#  สร้างข้อมูลจำลอง
X_synth, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=42)

#  Elbow Method
wcss_synth = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_synth)
    wcss_synth.append(kmeans.inertia_)

#  วาดกราฟ
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss_synth, marker='o', linestyle='--')
plt.title('Elbow Method for Synthetic Data')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```
