```python
print("1. การโหลดและตรวจสอบข้อมูลเบื้องต้น")
import pandas as pd
import io


with open('student.csv', 'r', encoding='utf-8') as f:
    raw_data = f.read()

lines = raw_data.strip().split('\n')
header = lines[0].split(',')
data = [line.split(',') for line in lines[1:] if line.strip()]

df = pd.DataFrame(data, columns=header)

for col in ["MidtermScore", "FinalScore", "ProjectScore", "Attendance"]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# แสดงข้อมูล 10 แถวแรก
print("ข้อมูล 10 แถวแรก:")
print(df.head(10))

# แสดงชนิดข้อมูลและตรวจสอบค่าว่าง
print("\nข้อมูลเบื้องต้น:")
df.info()

# แสดงคอลัมน์ที่มีข้อมูลขาด
print("\nจำนวนข้อมูลที่ขาดในแต่ละคอลัมน์:")
print(df.isnull().sum())
import pandas as pd

# แปลงคอลัมน์ตัวเลขให้เป็นชนิดที่เหมาะสม
numeric_columns = ['StudentID', 'MidtermScore', 'FinalScore', 'ProjectScore', 'Attendance']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # บังคับ NaN หากแปลงไม่ได้

# แสดงข้อมูลเบื้องต้น
print("ตรวจสอบชนิดข้อมูลและค่าว่างด้วย .info():")
df.info()

print("จากผลรับที่ได้ในข้อที่ 1")
print("คอลัมน์ที่มีข้อมูลขาดหายคือ MidtermScore (2 ค่า) และ FinalScore (1 ค่า)")
print("ชนิดข้อมูลเป็น float64 เนื่องจาก Pandas จะแปลงคอลัมน์ที่มีค่าว่างเป็น float โดยอัตโนมัติ")
```
