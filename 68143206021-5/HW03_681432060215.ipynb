```python
import pandas as pd
import numpy as np

#1.การโหลดและตรวจสอบข้อมูลเบื้องต้น
df = pd.read_csv('student_data.csv')
print(df.head(10))
df.info()
#คำตอบ จากผลลัพธ์ .info() คอลัมน์ที่มีข้อมูลหาย = "FinalScore" และ "MidtermScore" มีชนิดข้อมูล(Dtype)เป็น "int64"

#2.การจัดการข้อมูลที่ขาดหาย (Missing Value)
print(df.isnull().sum())

midtermscore_median = df["MidtermScore"].median()
finalscore_median = df["FinalScore"].median()

print(f'ค่ามัธยฐานของคะแนน Midterm: {midtermscore_median}')
print(df['MidtermScore'].fillna(midtermscore_median, inplace=True))

print(f'ค่ามัธยฐานของคะแนน Final: {finalscore_median}')
print(df['FinalScore'].fillna(finalscore_median, inplace=True))
print('\n')
print(df.isnull().sum())
print('\n' +'='*40+ '\n')

#3.แปลงชนิดข้อมูล (Data Type Conversion)
df['MidtermScore'] = df['MidtermScore'].astype(int)
df['FinalScore'] = df['FinalScore'].astype(int)
df.dtypes

#4. การส้ร้างคอลัมน์ใหม่จากข้อมูลเดิม
df['FullName'] = df['FirstName'] + ' ' + df['LastName']
print(df)

#5. การคำนาณคะแนนรวม
df['TotalScore'] = df['MidtermScore'] + df['FinalScore'] + df['ProjectScore']
print(df)

#6. การคำนวณการตัดเกรด
def calculate_grade(total_score):

    if total_score > 230:
        grade = 'A+'
    elif total_score >= 200:
        grade = 'A'
    elif total_score >= 170:
        grade = 'B'
    elif total_score >= 140:
        grade = 'C'
    elif total_score >= 100:
        grade = 'D'
    else:
        grade = 'F'    
    return grade

df['FinalGrade'] = df['TotalScore'].apply(calculate_grade)
print(df[['FinalGrade']].head(15))

#7. การคัดกรองข้อมูลตามเงื่อนไขที่ซับซ้อน
print('\n นักศึกษาสาขาวิศวกรรมคอมพิวเตอร์ (Computer Engineering) ที่ได้เกรด A')
computer_engineering_student_name = df[(df['Major'] == 'Computer Engineering') & (df['FinalGrade'] == 'A')]
print(computer_engineering_student_name)

print('\nนักศึกษาที่เข้าเรียนน้อยกว่า 70%')
student_name_attendance = df[df['Attendance'] < 70]
print(student_name_attendance)

#8. การเรียงลำดับข้อมูล
print('\nเรียงลำดับข้อมูลนักศึกษาทั้ง DataFrame ตาม TotalScore จาก มากไปน้อย และแสดง 5 อันดับแรก')
df.sorted = df.sort_values(by='TotalScore', ascending=False)
print(df.sorted.head())

#9. การจัดกลุ่มและสรุปผลข้อมูล (Grouping and Aggregation)
print(df.groupby('Major').agg({
    'TotalScore': 'mean',
    'FinalScore': 'mean'
}))

#10. การเตรียมและการบันทึกรายงานสรุป
report_df = df[['StudentID', 'FullName', 'Major', 'TotalScore', 'FinalGrade']]
report_df.to_csv('student_grade_report.csv')
```
