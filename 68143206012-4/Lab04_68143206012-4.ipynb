```python
import seaborn as sns
import matplotlib.pyplot as plt
# โหลดข้อมูล Seaborn
tips = sns.load_dataset('tips')
```

```python
# --- กราฟที่ 1: การกระจายตัวของยอดรวมใบเสร็จ ---
# สร้างพื้นที่สำหรับวาดกราฟ ขนาด 10x6 นิ้ว
plt.figure(figsize=(10, 6))

# สร้างกราฟฮิสโทแกรมด้วย Seaborn
# data=tips: ใช้ข้อมูลจาก DataFrame 'tips'
# x='total_bill': ใช้ข้อมูลจากคอลัมน์ 'total_bill' เป็นป็แกน X
# kde=True: วาดเส้น ประมาณความหนาแน่นน่ (Kernel DensityEstimate) ทับลงบนกราฟ
sns.histplot(data=tips, x='total_bill', kde=True)

# ตั้งชื่อหัวข้อของกราฟ
plt.title('Distribution of Total Bill')

# แสดงกราฟที่สร้างขึ้น
plt.show()
```

```python
# --- กราฟที่ 2: จํานวนลูกค้าในแต่ละวัน ---
# สร้าง Figure ใหม่สําหรับกราฟถัดไป
plt.figure(figsize=(8, 5))

# สร้างกราฟแท่งนับความถี่
# x='day': ใช้ข้อมูลจากคอลัมน์ 'day'
# order=[...]: กํา หนดลํา ดับของแท่งบนแกน X ให้เรียงตามที่เราต้องการ
sns.countplot(data=tips, x='day', order=['Thur','Fri','Sat','Sun'])

# ตั้งตั้ ชื่อหัวข้อของกราฟ
plt.title('Number of Customers per Day')

# แสดงกราฟ
plt.show()
```

```python
# --- กราฟที่ 3: การกระจายตัวของเงินทิป ---
# สร้าง Figure ใหม่อีกครั้ง
plt.figure(figsize=(10, 6))

# สร้างกราฟกล่อง (Box Plot) เพื่อพื่ดูการกระจายตัวและค่าผิดปกติ (Outliers)
# x='tip': ใช้ข้อมูลจากคอลัมน์ 'tip'
sns.boxplot(data=tips, x='tip')

# ตั้งชื่อหัวข้อของกราฟ
plt.title('Distribution of Tip Amount')

# แสดงกราฟ
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# โหลดข้อมูล Seaborn
titanic = sns.load_dataset('titanic')
```

```python
# --- กราฟที่ 1: อายุ vs ค่าโดยสาร ---
# สร้างพื้นที่สําหรับวาดกราฟ (Figure) ขนาด 10x6 นิ้ว
plt.figure(figsize=(10, 6))

# สร้างกราฟจุด (Scatter Plot) เพื่อพื่ดูความสัมพันธ์ระหว่างตัวแปรเชิงปริมาณ 2 ตัว
# x='age': กํา หนดแกน X เป็นอายุ
# y='fare': กํา หนดแกน Y เป็นค่าโดยสาร
# hue='survived': กํา หนดให้ "สี"สี ของจุดแตกต่างกันตามค่าในคอลัมน์'survived' (0=ตาย, 1=รอด)
sns.scatterplot(data=titanic, x='age', y='fare',hue='survived')

# ตั้งชื่อหัวข้อของกราฟ
plt.title('Relationship between Age, Fare, andSurvival')

# แสดงกราฟที่สร้างขึ้น
plt.show()
```

```python
# --- กราฟที่ 2: เพศ vs อัตราการรอดชีวิต ---
# สร้างพื้นที่สําหรับวาดกราฟ (Figure) ขนาด 8x5 นิ้ว
plt.figure(figsize=(8, 5))

# สร้างกราฟแท่ง (Bar Plot) เพื่อพื่ เปรียรีบเทียบค่าเฉลี่ย
# x='sex': แบ่งกลุ่มตามเพศ
# y='survived': ค่าที่นํามาคํานวณค่าเฉลี่ย (เนื่อนื่ งจากค่าคือ 0 กับ1 ค่าเฉลี่ยจึงจึเท่ากับอัตราการรอดชีวิชีตวิ )
sns.barplot(data=titanic, x='sex', y='survived')

# ตั้งตั้ ชื่อหัวข้อของกราฟ
plt.title('Survival Rate by Sex')

# เปลี่ยนชื่อชื่แกน Y ให้อ่านเข้าใจง่าย
plt.ylabel('Survival Rate')

# แสดงกราฟ
plt.show()
```

```python
# --- กราฟที่ 3: ชั้นโดยสาร vs อายุ ---
# สร้าง Figure ใหม่อีกครั้ง
plt.figure(figsize=(10, 6))

# สร้าง Box Plot เพื่อเปรียบเทียบการกระจายตัวของข้อมูลเชิงปริมาณในแต่ละกลุ่ม
# x='pclass': แบ่งกลุ่มตามชั้นโดยสาร
# y='age': แสดงการกระจายตัวของอายุในแต่ละกลุ่ม
sns.boxplot(data=titanic, x='pclass', y='age')

# ตั้งชื่อหัวหัข้อของกราฟ
plt.title('Age Distribution by Passenger Class')

# แสดงกราฟ
plt.show()
```

```python
# --- กราฟที่ 1: Pair Plot ---
# เลือกเฉพาะคอลัมน์ที่สนใจเพื่อพื่สร้าง Pair Plot
subset = titanic[['survived','pclass','age','fare','sex']]

# สร้างตารางของกราฟที่แสดงความสัมพันธ์ของทุกคู่ตัวแปร
# data=subset: ใช้ข้อมูลจาก DataFrame 'subset' ที่เราเลือกไว้
# hue='survived': แยกสีขสีองจุดและเส้นตามการรอดชีวิต
# diag_kind='kde': กํา หนดให้กราฟในแนวทแยง (การกระจายตัวของตัวแปรเดียว) เป็นกราฟ KDE
sns.pairplot(subset, hue='survived',diag_kind='kde')

# แสดงกราฟที่สร้างขึ้น
plt.show()
```

```python
# --- กราฟที่ 2: Heatmap ---
# สร้างสําเนาของ DataFrame เพื่อป้องกันการเปลี่ยนแปลงข้อมูลต้นฉบับ
titanic_corr = titanic.copy()

# แปลงข้อมูล 'sex' ที่เป็นข้อความให้เห้ป็นตัวเลข (0, 1) เพื่อให้สามารถนําไปคํานวณ correlation ได้
titanic_corr['sex'] = titanic_corr['sex'].map({'male': 0,'female': 1})

# คํานวณค่าสหสัมพันธ์ (Correlation Matrix) ระหว่างคอลัมน์ที่เลือก
corr_matrix = titanic_corr[['survived','pclass','age','fare','sex']].corr()

# สร้าง Figure ใหม่สําหรับ Heatmap
plt.figure(figsize=(10, 8))

# สร้าร้ง Heatmap จาก Correlation Matrix
# data=corr_matrix: ใช้ตาราง correlation ที่คํานวณไว้
# annot=True: แสดงตัวเลขค่า Correlation กํากับในแต่ละช่อง
# cmap='coolwarm': กําหนดชุดสี (สีร้อน=บวก, สีเย็น=ลบ)
# fmt='.2f': กําหนดรูปแบบของตัวเลขที่แสดงเป็นทศนิยม 2 ตํา แหน่ง
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',fmt='.2f')

# ตั้งชื่อหัวข้อของกราฟ
plt.title('Correlation Matrix of Titanic Features')

# แสดงกราฟ
plt.show()
```

```python
# Violin Plot
sns.violinplot(data=titanic,x='pclass', y='age', hue='survived',split=True)
plt.title('Age Distribution by Class and Survival')
plt.show()
```

```python
# สร้างตารางกราฟด้วย catplot
# x='sex', y='survived': คือความสัมพันธ์หลักที่สนใจในแต่ละกราฟย่อย
# col='pclass': สั่งให้สร้างกราฟแยก "คอลัมน์"โดยแบ่งตามค่าใน 'pclass'
# kind='bar': กํา หนดให้ชนิดของกราฟในแต่ละช่องเป็น bar plot
sns.catplot(data=titanic, x='sex',y='survived', col='pclass', kind='bar')
plt.show()
```
