```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
x_synth, y_teue  = make_blobs(n_samples=300, centers=4, cluster_std=0.70, random_state=0)
plt.figure(figsize=(10,6))
plt.scatter(x_synth[:,0],x_synth[:,1],s=50)
plt.title('synthetic data distribution')
plt.grid(True)
plt.grid(True)
plt.show()
```

```python
wcss_synth =[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init='k-means++',random_state=42)
    kmeans.fit(x_synth)
    wcss_synth.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss_synth, marker='o', linestyle='--')
plt.title('Elbow Method for Synthetic Data')
plt.xlabel('numbre of clusters (k)')
plt.ylabel('wcss')
plt.grid(True)
plt.show()
```
