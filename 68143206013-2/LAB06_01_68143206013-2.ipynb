```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X_data, Y_data = make_blobs(n_samples=300, centers=4, cluster_std = 0.7,random_state=0)
plt.figure(figsize=(10,6))
plt.scatter( X_data[:,0] , X_data[:,1], s = 50)
plt.title('Synthetic Data Distribution')
plt.grid(True)
plt.show()

test = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++',random_state=0)
    kmeans.fit(X_data)
    test.append(kmeans.inertia_)

plt.figure(figsize=(10,6))
plt.plot(range(1,11),test)
plt.xlabel('Elbow Method')
plt.ylabel('Error Rate')
plt.title('Error Rate vs. K Value for Synthetic Data')
plt.grid(True)
plt.show()

optimai = 4
kmeans = KMeans(n_clusters=optimai, init='k-means++',random_state=42)
ykmean = kmeans.fit_predict(X_data)
plt.figure(figsize=(12,8))
colors = ['red','blue','green','magenta','cyan']
for i in range(optimai):
    plt.scatter( X_data[ykmean == i,0] , X_data[ykmean == i,1], s = 100, c = colors[i], label = f'Cluster {i+1}')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Synthetic Data Distribution')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

```
