```python
import pandas as pd
import numpy as np
#1
print('\n' "#1" '\n')
df = pd.read_csv('student_data.csv')
print(df.head(10))
df.info()

#ข้อ1 ตอบ column MidtermScore 2ค่า ,FinalScore 1ค่า // type float64ทั้งคู่
#2
print('\n' "#2" '\n')
df[['MidtermScore','FinalScore']] = df[['MidtermScore','FinalScore']].fillna(df[['MidtermScore','FinalScore']].median())
print(df.isnull().sum())
#3
print('\n' "#3" '\n')
print("ProjectScore",df['ProjectScore'].dtypes)
print(df['ProjectScore'].astype(int))
#4
print('\n' "#4" '\n')
df['Fullname'] = df['FirstName'] +' '+ df['LastName']
#5
print('\n' "#5" '\n')
df['TotalScore'] = df['MidtermScore']  + df['FinalScore'] + df['ProjectScore']
df['gradescore'] = (df['MidtermScore'] * 0.3) + (df['FinalScore']* 0.4) + (df['ProjectScore']* 0.3)
def calculate_grade(gradescore):
    if gradescore >= 80:
        return 'A'
    elif gradescore >= 70:
        return 'B'
    elif gradescore >= 60:
        return 'C'
    elif gradescore >= 50:
        return 'D'
    elif gradescore < 50:
        return 'F'
#6
print('\n' "#6" '\n')
df['Grade'] = df['gradescore'].apply(calculate_grade)
print(df.head(10))
#7
print('\n' "#7" '\n')
Computer_Engineer = df[(df['Major'] == 'Computer Engineering') & (df['Grade'] == 'A')]
print(Computer_Engineer[['Fullname', 'Major','TotalScore' , 'Grade']])
name_ = df[df['Attendance'] < 70]
print(name_[['Fullname', 'Major', 'Attendance']])
#8
print('\n' "#8" '\n')
dftotalscore = df.sort_values(by= 'TotalScore', ascending=False)
showtop5 = dftotalscore.head(5)
print(showtop5[['Fullname', 'Major', 'TotalScore', 'Grade']])
#9
print('\n' "#9" '\n')
dfgroup = df.groupby('Major').agg({'TotalScore' : 'mean' , 'FinalScore' : 'mean'})
#10
print('\n' "#10" '\n')
dfreport = df[[ 'StudentID' , 'Fullname' , 'Major' , 'TotalScore' , 'Grade']]
print (dfreport.head())
op_name = 'student_grade_report.csv'
dfreport.to_csv(op_name, index=False)
```
