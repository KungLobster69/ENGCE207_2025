```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X_synth, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.70, random_state=0)

plt.Figure(figsize=(10, 6))
plt.scatter(X_synth[:, 0], X_synth[:, 1],s=50)
plt.title("Synthetic Data")
plt.grid(True)
plt.show()
from sklearn.cluster import KMeans

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_synth)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X_synth)

plt.scatter(X_synth[:, 0], X_synth[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
s=20, c='red', marker='x') 
plt.title("K-Means Clustering Result")
plt.grid(True)
plt.show()
```
