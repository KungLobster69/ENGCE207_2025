```python
# 1. 
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 2. 
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.70, random_state=0)

# 3.  Scatter plot 
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c='blue', s=50, edgecolor='k')
plt.title("make_blobs")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
#  inertia  1  10
inertias = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)  #  X 
    inertias.append(kmeans.inertia_)

#  Elbow
plt.figure(figsize=(10,6))
plt.plot(k_range, inertias, marker='o')
plt.xlabel('Group (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method Appropriate number of groups')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

#  KMeans
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', edgecolors='black')
plt.title("Grouping results K-Means")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
#--------------------------------------------------------------
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='tab10')
plt.title("Real data group (y_true)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
